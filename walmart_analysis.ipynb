```python
walmart_store_grouped = walmart_df.groupby('Store') # grouping it by store for later use
```

```python
store_sums = walmart_store_grouped.agg({
    "Weekly_Sales" : "sum"
}).round(2) # getting total sum of each store
```

```python
store_sums.head(5)
```

```python
# let's find the most top ten revenue producers
store_sums.sort_values(by="Weekly_Sales", ascending=False, inplace=True)
store_sums.head(5)
```

```python
# store 20 produces the most revenue in the dataset recorded by a large margin with store 4 following suit.
```

```python
(store_sums.iat[0,0] - store_sums.iat[-1,-1]).round(2) # the difference between the highest revenue producing branch and lowest.
```

```python
# speaking of best performing, lets find out how they did over the course of the dataset
```

```python
walmart_df[walmart_df.Store == 20]
```

```python
walmart_df[walmart_df.Store == 20].Weekly_Sales.mean().round(2)
```

```python
# over the course of the measured range, the top performing store averaged an amount of two million in revenue
```

```python
# time to visualize
```

```python
top_ten_stores = store_sums.head(10)
```

```python
top_ten_stores.iat[0,0]
```

```python
# graph of highest performing stores
fig, ax = plt.subplots(figsize=(10,8)) # setting the general size
top_ten_graph = sns.barplot(x=top_ten_stores.index, y="Weekly_Sales", data=top_ten_stores, ax=ax)
top_ten_graph.set(xlabel="Store", ylabel="Weekly Sales in Hundred Millions", title="Top 10 Store Revenue")
```

The graph above shows how each store contributes above 200 hundred million in revenue.

```python
# distribution of revenue across all stores
fig, ax = plt.subplots(figsize=(10,8))
revenue = sns.barplot(data=store_sums, x=store_sums.index, y="Weekly_Sales", ax=ax)
revenue.set(title="Revenue for all Stores", xlabel="Store", ylabel="Weekly Sales in Hundred Millions")
```

for reference, this graph shows the revenue across all stores. This paints a clear picture of the top ten performing stores proving itself far more profitable than majority.

```python
unemployed_avg = walmart_store_grouped.agg({
    "Unemployment" : "mean"
}).round(3) # saving to new variable for later
```

```python
unemployed_avg # checking that it worked
```

```python
# graph of unemployment rate for each store location over the range of the dataset
fig, ax = plt.subplots(figsize=(10,8))
unemployed_avg_graph = sns.histplot(data=unemployed_avg, x="Unemployment", ax=ax)
unemployed_avg_graph.set(title="Unemployment rate", xlabel="Unemployment", ylabel="% of Unemployed")
```

Unemployment shows a larger trend toward the lower amounts as opposed to the higher amounts.

```python
fuel_avg = walmart_df.groupby("Date").agg({
    "Fuel_Price" : "mean"
}) # grouping by date for accuracy. We want average fuel behaviour per date across stores
```

```python
# next we want the fuel price over time
fig, ax = plt.subplots(figsize=(10,8))
fuel_price_graph = sns.lineplot(data=fuel_avg, x=fuel_avg.index, y="Fuel_Price", ax=ax)
fuel_price_graph.set(title="Fuel Price over Time", xlabel="Date", ylabel="Fuel Price")
```

2011 saw a massive increase in fuel prices which then lowered before fluctuating again near the same high. As this graph continues, the fuel prices begin to stabilize but not to lower back to its original price value. This confirms that products will continue to rise in prices due to transportation costs.

```python
walmart_df_date_group_sum = walmart_df.groupby('Date').agg({
    "Weekly_Sales" : "sum"
}) # we want weekly sales for each store over time
```

```python
walmart_df_date_group_sum
```

```python
# total sum of sales over time
fig, ax = plt.subplots(figsize=(10,8))
sales_over_time = sns.lineplot(data=walmart_df_date_group_sum, x="Date", y="Weekly_Sales", ax=ax)
sales_over_time.set(title="Weekly Sales Over Time", xlabel="Date", ylabel="Weekly Sales")
```

The data above shows us that Walmart's revenue consistently rises only in the december season or the last quarter of the year. This can be attributed to seasonal pricing and holiday festivities.
